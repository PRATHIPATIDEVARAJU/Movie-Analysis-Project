```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Duration (minutes)'], df['IMDb Rating'], color='darkblue', alpha=0.6)
plt.xlabel('Duration (minutes)')
plt.ylabel('IMDb Rating')
plt.title('IMDb Rating vs Duration')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['MetaScore'], bins=20, color='purple', edgecolor='black')
plt.xlabel('MetaScore')
plt.ylabel('Frequency')
plt.title('Distribution of MetaScores')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Year'], df['IMDb Rating'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('IMDb Rating')
plt.title('IMDb Ratings by Year')
plt.show()
```

```python
genre_counts = df['Genre'].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(genre_counts.index, genre_counts, color='seagreen')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Count of Movies by Genre')
plt.show()
```

```python
# Calculate average IMDb Rating per Genre
genre_ratings = df.groupby('Genre')['IMDb Rating'].mean().sort_values()

plt.figure(figsize=(12, 8))
genre_ratings.plot(kind='barh', color='skyblue')
plt.title('Average IMDb Rating by Genre')
plt.xlabel('Average IMDb Rating')
plt.ylabel('Genre')
plt.show()
```

```python
# Calculate average IMDb Rating per Year
year_ratings = df.groupby('Year')['IMDb Rating'].mean()

plt.figure(figsize=(12, 6))
year_ratings.plot(kind='line', marker='D', color='purple')
plt.title('Average IMDb Rating Over Years')
plt.xlabel('Year')
plt.ylabel('Average IMDb Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.boxplot(column='IMDb Rating', by='Certificates')
plt.title('IMDb Rating by Certificate')
plt.xlabel('Certificates')
plt.ylabel('IMDb Rating')
plt.show()
```

```python
# Calculate genre distribution
genre_counts = df['Genre'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired(range(len(genre_counts))))
plt.title('Genre Distribution')
plt.show()
```

```python
Certificates_counts = df['Certificates'].value_counts()

plt.figure(figsize=(12, 8))
plt.pie(Certificates_counts, labels=Certificates_counts.index, autopct='%1.1f%%', startangle=120)

plt.title('Distribution of Certificates')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
df.boxplot(column='Duration (minutes)', by='Genre')
plt.title('Duration by Genre')
plt.xlabel('Genre')
plt.ylabel('Duration (minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Duration (minutes)'], df['IMDb Rating'], alpha=0.5, color='green')
plt.title('Duration vs. IMDb Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('IMDb Rating')
plt.grid(True)
plt.show()
```

```python
# Calculate average MetaScore per Year
year_metacore = df.groupby('Year')['MetaScore'].mean()

plt.figure(figsize=(12, 6))
year_metacore.plot(kind='line', marker='o', color='orange')
plt.title('Average MetaScore Over Years')
plt.xlabel('Year')
plt.ylabel('Average MetaScore')
plt.grid(True)
plt.show()
```
